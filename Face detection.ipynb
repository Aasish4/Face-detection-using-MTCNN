# FACE DETECTION USING MTCNN

```python
# load the required dependencies
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle

from mtcnn.mtcnn import MTCNN
```

```python
# functino to draw bbox
def draw_box(filename, faces_list):
    
    # load the image
    data = pyplot.imread(filename)
    
    # display the image
    pyplot.imshow(data)
    
    # get the current axis
    ax = pyplot.gca()
    
    for face in faces_list:
        x, y, w, h = face['box']
        rect = Rectangle((x, y), w, h, fill=False, color='green')
        ax.add_patch(rect)
        
        # keypoint ploting
        for key, value in face['keypoints'].items():
            dot = Circle(value, radius=2, color='red')
            ax.add_patch(dot)
    
    # save the figure
    pyplot.savefig('output.jpg')
    pyplot.show()
    
```

```python
# input image path
filename = 'image.jpg'
```

```python
# read the image 
pixels = pyplot.imread(filename)
```

```python
# initiate the detector constructor
detector = MTCNN()
```

```python
# detect face on the given image
faces = detector.detect_faces(pixels)
```

```python
# draw the bounding box and keypoints
draw_box(filename, faces)
```
